```python
import seaborn as sns
import matplotlib.pyplot as plt

import distributions
```

```python
total_budget = 2400
n_intervals = 12
```

```python
def create_plot(ax, P, method):
    color = '#6BAED6' if method == 'linear' else '#74C476'
    values = distributions.distribute_budget_linear(P, total_budget, n_intervals) if method == 'linear' else distributions.distribute_budget_exponential(P, total_budget, n_intervals)
    sns.barplot(x=list(range(len(values))), y=values, color=color, ax=ax)
    print(f'{method}: {values}')
    ax.set_title(f'P={P}, {method}')
    ax.set_xlabel('Week')
    ax.set_ylabel('Budget')
```

```python
def create_plots(P):
    fig, axes = plt.subplots(2, 1, figsize=(13, 8))
    create_plot(axes[0], P, 'linear')
    create_plot(axes[1], P, 'exponential')
    plt.tight_layout()
    plt.show()
```

Check distributions if P parameter id -1. In this case all the budget is for the last month.

```python
create_plots(-1)
```

P = -0.7. At this parameter value, the distribution of values resembles linear and exponential distributions, where a significant portion of the budget is allocated in the last months.

```python
create_plots(-0.7)
```


P = -0.5.
At this parameter value, the values of the linear distribution resemble an increasing arithmetic progression, while the values of the exponential distribution resemble an increasing geometric progression, where a significant portion of the budget is allocated in the last months.

```python
create_plots(-0.5)
```

P = -0.2.
